### 1. Calculate $p$, $p^*$, $\text{error}$, $L$

```python
import torch

# Input features
x = torch.tensor([1.0, 0.0, 1.0])   # sunny, long queue, healthy

# Initial weights and bias
weights = torch.tensor([4.0, 3.0, 4.0], requires_grad=True)
bias = torch.tensor(-4.0, requires_grad=True)

# True label p*
p_star = torch.tensor(1.0)   # yes, we want the output to be 1

# Forward pass
z = torch.dot(weights, x) + bias
p = torch.sigmoid(z)

# Error and loss
error = p - p_star
loss = (p - p_star) ** 2

print(f"z = {z.item():.4f}")
print(f"p (predicted probability) = {p.item():.4f}")
print(f"p* (true value) = {p_star.item():.4f}")
print(f"error = p - p* = {error.item():.4f}")
print(f"loss = {loss.item():.8f}")
```

### 2. Using gradient descent to update the weights

```python
loss_threshold = 0.0001
max_epochs = 10000   

epoch = 0
lr = 0.1

while True:
    epoch += 1

    # Forward pass
    z = torch.dot(weights, x) + bias
    p = torch.sigmoid(z)
    loss = (p - p_star) ** 2

    # Stop condition
    if loss.item() < loss_threshold or epoch > max_epochs:
        print("Training stopped.")
        break

    # Clear old gradients
    if weights.grad is not None:
        weights.grad.zero_()
    if bias.grad is not None:
        bias.grad.zero_()

    # Backward pass
    loss.backward()

    # Update
    with torch.no_grad():
        weights -= lr * weights.grad
        bias -= lr * bias.grad

    print(
        f"Epoch {epoch:2d} | "
        f"p = {p.item():.4f} | "
        f"loss = {loss.item():.8f} | "
        f"weights = {weights.detach().numpy()} | "
        f"bias = {bias.item():.4f}"
    )
```

## Talking points

### How to find the value of $p$, and compare it to $p^*$

For a single-layer perceptron, we first compute the weighted sum:
$$z = w_1x_1 + w_2x_2 + w_3x_3 + b$$
Then we apply the sigmoid activation function to get the predicted probability:
$$p = \sigma(z) = \frac{1}{1 + e^{-z}}$$

- $p$: predicted probability
- $p^*$: true target value (ground truth)

We compare them by computing:

$$\text{error} = p - p^*$$

If $p$ is close to $p^*$, the model is doing well.
If the difference is large, the model needs to adjust its weights.

### How to compute the loss function

We can measure the error using a loss function.
A simple choice is the **Mean Squared Error (MSE)**:
$$L = (p - p^*)^2$$
This tells us how far the prediction is from the true value.

### How Gradient Descent updates the weights

Gradient Descent updates each weight in the direction that reduces the loss:

$$w_j = w_j - \eta \frac{\partial L}{\partial w_j}$$

where:

- $w_j$ is a weight
- $\eta$ is the learning rate
- $\frac{\partial L}{\partial w_j}$ is the gradient of the loss with respect to the weight

This means we compute how much each weight contributes to the error, then slightly adjust it to reduce the error.

### What is **Gradient Descent**

**Gradient Descent** is an optimization algorithm used in machine learning and deep learning to minimize a **loss function** by iteratively updating the model’s parameters.

In most machine learning models, such as linear regression or neural networks, the model contains parameters (e.g., weights and biases) that determine how inputs are transformed into predictions. During training, the goal is to find parameter values that produce predictions as close as possible to the true target values. This is measured using a **loss function**, which quantifies the error between predicted and actual outputs.

Gradient Descent works by calculating the gradient of the loss function with respect to the model parameters. The gradient represents the direction of the steepest increase of the loss function. To reduce the loss, the algorithm moves the parameters in the opposite direction of the gradient.

The training process typically follows these steps:

- Forward pass: The model computes predictions from the input data.

- Loss computation: The loss function measures the difference between predictions and true values.

- Backpropagation: The gradient of the loss with respect to each parameter is calculated.

- Parameter update: The parameters are adjusted using the Gradient Descent update rule.

This process is repeated many times until the loss function converges to a minimum or reaches an acceptable level.

### How do we use Gradient Descent to do Back Propagation

Gradient Descent and Backpropagation work together during the training of neural networks. Backpropagation computes the gradients of the loss function with respect to the model parameters, and Gradient Descent uses those gradients to update the parameters in order to minimize the loss.

First, during the forward pass, the input data is passed through the neural network layer by layer to produce a prediction. The model then compares the prediction with the true target value using a loss function, which measures the prediction error.

Next, backpropagation is applied. Backpropagation uses the chain rule from calculus to compute how the loss changes with respect to each parameter (weights and biases) in the network. It starts from the output layer and propagates the error backward through the network, calculating the gradient of the loss with respect to each parameter.

Once these gradients are computed, Gradient Descent updates the parameters. The parameters are adjusted in the opposite direction of the gradient because the gradient indicates the direction of the steepest increase of the loss. By moving in the opposite direction, the algorithm reduces the loss.
